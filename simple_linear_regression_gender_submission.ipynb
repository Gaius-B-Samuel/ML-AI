```python
plt.scatter(x=df['PassengerId'], y=df['Survived'])
plt.show()
```

```python
X = df[['PassengerId']]
y = df[['Survived']]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
lr=LinearRegression()
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
print(mean_squared_error(y_pred, y_test))
```
