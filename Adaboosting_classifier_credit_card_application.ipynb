```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import confusion_matrix
import numpy as np
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
data = pd.read_csv("/content/drive/MyDrive/ML & AI/Credit_Card_Applications (1).csv")
```

```python
data
```

```python
#data preparation

training_set, test_set = train_test_split(data, test_size = 0.2, random_state = 1)

X_train = training_set.iloc[:,1:15]
Y_train = training_set.iloc[:,15]
X_test = test_set.iloc[:,1:15]
Y_test = test_set.iloc[:,15]
```

```python
X_train
```

```python
adaboost = AdaBoostClassifier(n_estimators=50, base_estimator= None,learning_rate=1.5, random_state = 1)
adaboost.fit(X_train,Y_train)
Y_pred = adaboost.predict(X_test)
```

```python
cm = confusion_matrix(Y_test,Y_pred)
accuracy = np.mean(Y_pred==Y_test)
print("\nAccuracy Of AdaBoost For The Given Dataset :", accuracy)
```

```python
print('\nAccuracy Of AdaBoost For The Given Dataset: {:.2f}'.format(accuracy))
```
