```python
acc=[]
for i in range(3,50,2):
  modelknn=KNC(n_neighbors=i)
  modelknn.fit(trainx,trainy)
  predknn=modelknn.predict(testx)
  knnacc=np.mean(predknn==testy)
  acc.append([knnacc,i])
```

**LOGISTIC REGRESSION ALGORITHM**

```python
#model building using Logistic Regression Algorithm
logistic_regression=LogisticRegression()
logistic_regression.fit(trainx,trainy)
y_pred=logistic_regression.predict(testx)
acc=np.mean(y_pred==testy)
```

```python
acc
```

**DECISION TREE ALGORITHM**

```python
#model buidling using Decision Tree Algorithm
modelDT=DT(criterion="entropy")
modelDT.fit(trainx,trainy)
predDT=modelDT.predict(testx)
DTacc=np.mean(predDT==testy)
```

```python
DTacc
```

**RANDOM FOREST ALGORITHM**

```python
#model building using Random Forest Algorithm (3001 is the number of columns)
modelRFC=RFC(n_jobs=3001,n_estimators=50,oob_score=True,criterion="entropy")
modelRFC.fit(trainx,trainy)
predRFC=modelRFC.predict(testx)
RFCacc=np.mean(predRFC==testy)
```

```python
RFCacc
```

**SUPPORT VECTOR MACHINE** **- Linear Method**

```python
#model buidling using SVM Algorithm (Method 1 - Linear)
model_linear=SVC(kernel="linear")
model_linear.fit(trainx,trainy)
pred_test_linear=model_linear.predict(testx)
acclinear=np.mean(pred_test_linear==testy)
```

```python
acclinear
```

**SUPPORT VECTOR MACHINE** **- Polynomial Method**

```python
#model buidling using SVM Algorithm (Method 2 - Polynomial)
model_poly=SVC(kernel="poly")
model_poly.fit(trainx,trainy)
pred_test_poly=model_poly.predict(testx)
accpoly=np.mean(pred_test_poly==testy)
```

**SUPPORT VECTOR MACHINE** **- RBF Method**

**SUPPORT VECTOR MACHINE** **- Sigmoid Method**

```python
#model buidling using SVM Algorithm (Method 4 - Sigmoid)
model_sigmoid=SVC(kernel="sigmoid")
model_sigmoid.fit(trainx,trainy)
pred_test_sigmoid=model_sigmoid.predict(testx)
accsigmoid=np.mean(pred_test_sigmoid==testy)
```
